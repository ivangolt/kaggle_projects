```python
def handle_outliers_all_columns(dataframe, low_quantile=None, high_quantile=None):
    """
          .

    :
    - dataframe: pandas DataFrame,  
    - lower_bound: float,      ( None,  )
    - upper_bound: float,      ( None,  )

    :
    -        .
    """
    #  ,    
    df = dataframe.copy()

    #    
    numeric_columns = df.select_dtypes(include=['number']).columns.to_list()
    
    #      
    for column_name in numeric_columns:
        lower_bound = df[column_name].quantile(low_quantile)
        higher_bound = df[column_name].quantile(high_quantile)
        df = df[(df[column_name] >= lower_bound) & (df[column_name] <= higher_bound)]
        
        
    return df
```

```python
train_df = handle_outliers_all_columns(train_df, low_quantile=0.01, high_quantile=0.99)
```

# Pipeline processing

```python
num_cols.remove('SalePrice')
```

```python
y_train = train_df['SalePrice']
X_train = train_df.drop(columns=['SalePrice'], axis=1)
X_test = test_df
y_test = test_target


num_pipeline = Pipeline(steps=[('nan_imputer_num', SimpleImputer(missing_values=np.nan, strategy='median')),
                                ('num_scaler', MinMaxScaler())])

ohe_pipeline = Pipeline(steps=[('nan_imputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
                               ('ohe_scaler', OneHotEncoder(handle_unknown='ignore',
                                                           drop='first',
                                                           sparse_output=False))])

data_preprocessing = ColumnTransformer(transformers=[
    ('ohe', ohe_pipeline, ohe_cols),
    ('num', num_pipeline, num_cols)],
    remainder='passthrough')


model_pipeline = Pipeline(steps=[('data_preprocessor', data_preprocessing),
                                 ('models', XGBRegressor())])
```

```python
model_pipeline.fit(X_train, y_train)
```

```python
y_predict = model_pipeline.predict(X_test)
```

```python
r2_score(y_pred=y_predict, y_true=y_test)
```

```python
mean_squared_error(y_pred=y_predict, y_true=y_test, squared=False)
```

```python
cross_val_score(model_pipeline, X=X_train, y=y_train, cv=10, scoring='r2')
```
